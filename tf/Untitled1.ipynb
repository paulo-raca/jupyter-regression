```python
import pandas as pd
import sympy
from data_fit import DataFit
import sympy.physics.units as units
import tensorboard
```

```python
datafit = DataFit()
datafit.add_experimental_data("x", units.meter, "Displacement")
datafit.add_experimental_data("F_exp", units.newton, "Measured Force")
datafit.add_parameter("k", units.newton/units.meter, "Sprint constant")
datafit.add_expr("F", units.newton, datafit.k * datafit.x, "Predicted force")
datafit.add_optimization_objective("F_MSE", units.newton, datafit.F_exp, datafit.F)
data = pd.DataFrame(data=dict(
    F_exp=[1,2,3,4], x=[1.1,2.1,3.2,4.1]
))
datafit.fit(data)
```

```python
datafit = DataFit()
datafit.add_experimental_data("Y_exp", 1, "out")
datafit.add_experimental_data("t", 1, "in")
datafit.add_parameter("Y_0", 1, "Initial value of Y")
datafit.add_parameter("k", 1, "some factor")
datafit.add_differential_expr("Y", 1, datafit.k, datafit.t, datafit.Y_0, "Estimated Y")
datafit.add_optimization_objective("F_MSE", 1, datafit.Y_exp, datafit.Y)
data = pd.DataFrame(data=dict(
    Y_exp=[4.,8,16,32,64,128,256], t=[0.,1,2,3,4,5,6]
))
datafit.fit(data, 10)
#print()
```
